```python
import pdfplumber

# 数据库 提取 线 划线 文本行
raw = r".\input\自定义生成测试 .pdf"
pdf = pdfplumber.open(raw)
first = pdf.pages[0]  # 肯定只有一页
# first = first.within_bbox((26, 49, 410, 577))
a = []
b = []
c=[]
for i in first.lines:
    if (i["width"] < 20) and (i["width"] > 5):
        a.append(i)
    elif i["width"] > 30:
        b.append(i)
    elif i["height"]>5:
        c.append(i)
im = first.to_image(resolution=200)
im.draw_lines(a, stroke_width=6, stroke="blue")
im.draw_lines(b, stroke_width=6, stroke=(241, 161, 7))
im.draw_lines(c, stroke_width=6, stroke=(225, 60, 225))
im.draw_rects(first.extract_text_lines(), stroke="green")
im.draw_rects(first.extract_words(), stroke="red")
im.save(r".\output\text & line.png")
```

```python
import pdfplumber

# 航路点 输出 文本行 单词 文字位置
raw = r"F:\航图\中国民航国内航空资料汇编 NAIP 2402\terminals\ZBAA-4Y02.pdf"
pdf = pdfplumber.open(raw)
first = pdf.pages[0]  # 肯定只有一页
first = first.crop((26, 67, 410, 560))
a=[]
for i in first.lines:
    if (i["width"] < 3) and (i["height"] > 20):
        a.append(i)
im = first.to_image(resolution=200)
im.draw_lines(a, stroke_width=5, stroke="blue")
im.draw_rects(first.extract_text_lines(), stroke="green")
im.draw_rects(first.extract_words(), stroke="red")
im.save(r".\output\word.png")
im.show()
# 输出提取的行文本
texts = first.extract_text_lines()
text_only = [i["text"] for i in texts]
for i in text_only:
    print(i)
a = []
chars = [i["chars"] for i in texts]
for iChar in chars:
    for info in iChar:
        print("({},{})".format(info["text"], round(info["x0"], 1)), end=' ')
    print('\n------------')
```

```python
import pdfplumber

# 数据库 每行文本(split()后的)
raw = r"D:\Python项目\fenix数据\input\ZULS-4D.pdf"
txt = open(r"output\table.txt", 'w', encoding="utf-8")
pdf = pdfplumber.open(raw)
first = pdf.pages[0]
a = first.extract_table()
for i in range(len(a)):
    print(a[i])
    a[i] = list(map(lambda x: x if x is not None else '', a[i]))
    a[i] = list(map(lambda x: x.replace("¡ã", '°'), a[i]))
    a[i] = '*'.join(a[i]) + '\n'
txt.writelines(a)
txt.close()
```
